```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set palette to viridis
plt.style.use("seaborn-v0_8")
```

## Influences

The `.npy` files are saved after each training run.

```python
nsa_influences = np.log10(np.load("artifacts/influences_tensor_nsa.npy"))
vanilla_influences = np.log10(np.load("artifacts/influences_tensor_erwin.npy"))
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(
    nsa_influences,
    label="Erwin + NSA (N influenced: 3586)",
    linewidth=2,
    bins=50,
    edgecolor="black",
    alpha=0.7,
    density=True,
)
ax.hist(
    vanilla_influences,
    label="Erwin (N influenced: 3586)",
    linewidth=2,
    bins=50,
    edgecolor="black",
    alpha=0.7,
    density=True,
)
ax.legend(fontsize=20)
ax.set_xlabel("log10(Gradient)", fontsize=20)
ax.set_ylabel("Normalized Count", fontsize=20)
ax.tick_params(axis="both", which="major", labelsize=20)
ax.tick_params(axis="both", which="minor", labelsize=18)
ax.set_title("Influence of One Node over Point Cloud", fontsize=24)
```
